```python
URL = "https://labs.perplexity.ai/?ref=maginative.com"
s = Service('D:/chromedriver_win32/chromedriver.exe')
text_area = '#__next > main > div > div > div.sticky.bottom-0.border-t.z-20.border-borderMain\/60.dark\:border-borderMainDark\/80.divide-borderMain\/60.dark\:divide-borderMainDark\/80.ring-borderMain.dark\:ring-borderMainDark.bg-background.dark\:bg-backgroundDark > div:nth-child(2) > div > div > div > div > div > div > div > div.w-full.outline-none.focus\:outline-none.focus\:ring-borderMain.font-sans.flex.items-center.dark\:bg-offsetDark.dark\:text-textMainDark.dark\:placeholder-textOffDark.dark\:border-borderMainDark.dark\:focus\:ring-borderMainDark.selection\:bg-superDuper.selection\:text-textMain.duration-200.transition-all.bg-background.border.text-textMain.border-borderMain.focus\:ring-1.placeholder-textOff.rounded-t-\[32px\].rounded-b-\[32px\].py-md.px-lg.pr-\[49px\].md\:pr-\[59px\] > textarea'
send_button = '#__next > main > div > div > div.sticky.bottom-0.border-t.z-20.border-borderMain\/60.dark\:border-borderMainDark\/80.divide-borderMain\/60.dark\:divide-borderMainDark\/80.ring-borderMain.dark\:ring-borderMainDark.bg-background.dark\:bg-backgroundDark > div:nth-child(2) > div > div > div > div > div > div > div > div.absolute.flex.items-center.gap-sm.bg-background.dark\:bg-offsetDark.rounded-full.right-sm > button > div > svg > path'
copy_button = '#__next > main > div > div > div.grow > div > div.h-full.flex.flex-col > div > div > div > div:nth-child(3) > div > div > div.text-right > div.flex.items-start.mt-xs.gap-x-xs > button > div.flex.items-center.leading-none.justify-center.gap-xs > div'
```

```python
# driver = webdriver.Chrome(service=s)
# driver.get(URL)
# driver.find_element(By.CSS_SELECTOR, text_area).send_keys(news_text + '\n')
# driver.find_element(By.CSS_SELECTOR, copy_button).click()
```

```python
def crawl(index):
    driver = webdriver.Chrome(service=s)
    driver.get(URL)
    news_text = f"Summarize the following in 300 letters.[{df.loc[index].article[:3500]}]"
    driver.find_element(By.CSS_SELECTOR, text_area).send_keys(news_text + '\n')
    time.sleep(6)
    try:
        driver.find_element(By.CSS_SELECTOR, copy_button).click()
    except:
        pass
    driver.close()
    # return clipboard.paste()
```

```python
def crawl_time_limit(index, time_limit=40):
    p = multiprocessing.Process(target=crawl(index))
    p.start()
    p.join(time_limit)
    if p.is_alive():
        clipboard.copy('')
        p.terminate()
        p.join()        
```

```python
df = pd.read_csv('./df_mistral.csv')
```

```python
idx = df[df.summary.isnull()].index
idx
```

```python
for i in tqdm(idx):
    try:
        crawl_time_limit(i)
        df.summary.loc[i] = clipboard.paste()
        df.to_csv('./df_mistral.csv', index=False)
    except:
        driver.close()
```
